```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
from nltk.tokenize import RegexpTokenizer

token_dict = {}
alltext = [P1, P2, P3, D1, D2, D3, C1, C2, C3, R1, R2, R3]

def stem_tokens(tokens, stemmer):
    stemmed = []
    for item in tokens:
        stemmed.append(stemmer.stem(item))
    return stemmed

def tokenize(text):
    tk = RegexpTokenizer(r'((?<=[^\w\s])\w(?=[^\w\s])|(\W))+', gaps=True)
    tokens = tk.tokenize(text)
    return tokens

for i,text in enumerate(alltext):
    lowers = text.lower()
    no_punctuation = lowers.translate(string.punctuation)
    token_dict[book_id[i]] = no_punctuation

tfidf = TfidfVectorizer(tokenizer=tokenize, stop_words='english',ngram_range =(2,2))
tfs = tfidf.fit_transform(token_dict.values())
# tfidf.transform([P1])
```

```python
str = P1
response = tfidf.transform([str])
# print(response)
feature_names = tfidf.get_feature_names()
r = []
for col in response.nonzero()[1]:
    if feature_names[col].count(' ')==1:
        bi = feature_names[col].split(' ')
        if '-' not in bi and '(' not in bi and ',' not in bi and '“' not in bi and '‘' not in bi and '’' not in bi:
            r.append([bi, response[0, col]])
sorted(r,key=lambda x:x[1],reverse=True)
```

```python
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import subjectivity
from nltk.sentiment import SentimentAnalyzer
from nltk.sentiment.util import *
n_instances = 5000
train_instances = int(0.8 * n_instances)
subj_docs = [(sent, 'subj') for sent in subjectivity.sents(categories='subj')[:n_instances]]
obj_docs = [(sent, 'obj') for sent in subjectivity.sents(categories='obj')[:n_instances]]
```

```python
train_subj_docs = subj_docs[:train_instances]
test_subj_docs = subj_docs[train_instances:n_instances]
train_obj_docs = obj_docs[:train_instances]
test_obj_docs = obj_docs[train_instances:n_instances]
training_docs = train_subj_docs+train_obj_docs
testing_docs = test_subj_docs+test_obj_docs
sentim_analyzer = SentimentAnalyzer()
all_words_neg = sentim_analyzer.all_words([mark_negation(doc) for doc in training_docs])
```

```python
unigram_feats = sentim_analyzer.unigram_word_feats(all_words_neg, min_freq=4)
sentim_analyzer.add_feat_extractor(extract_unigram_feats, unigrams=unigram_feats)
```

```python
training_set = sentim_analyzer.apply_features(training_docs)
test_set = sentim_analyzer.apply_features(testing_docs)
```

```python
trainer = NaiveBayesClassifier.train
classifier = sentim_analyzer.train(trainer, training_set)
for key,value in sorted(sentim_analyzer.evaluate(test_set).items()):
    print('{0}: {1}'.format(key, value))
```

```python
sent_all = []
for i in alltext:
    sent = sent_tokenize(i)
    ss = np.array([0.0]*4)
    c = 0
    for sentence in sent:
        if len(sentence) >1:
            ss += np.array(list(sid.polarity_scores(sentence).values()))  # c - + n
            c += 1
    sent_all.append((ss/c).tolist())
```

```python
sent_matrix = pd.DataFrame(sent_all,index=book_id,columns = ['Compound','Negative','Positive','Neutral'])
sent_matrix['author'] = ['Edgar Allan Poe']*3 + ['Arthur Conan Doyle'] * 3 + ['Agatha Christie'] * 3 + ['Raymond Chandler'] * 3
```
